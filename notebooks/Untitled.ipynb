```python
import sys
sys.path.append("..")


import numpy as np
import pandas as pd


# dependence measures
from pysim.kernel.hsic import HSIC

# MATPLOTLIB Settings
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# SEABORN SETTINGS
import seaborn as sns
sns.set_context(context='talk',font_scale=0.7)

%load_ext autoreload
%autoreload 2
```

```python
# random state
seed = 123
rng = np.random.RandomState(seed)

# generate fake data
n_samples = 100

X = rng.rand(n_samples)
Y = 10 * X + 0.1 * rng.randn(n_samples)

sns.scatterplot(x=X, y=Y);
```

```python
from sklearn.metrics.pairwise import rbf_kernel as rbf_sklearn
```

```python
hsic_clf = HSIC(
    kernel_X="rbf", kernel_Y="rbf",
    kernel_params_X={'gamma': 1.0,},
    kernel_params_Y={'gamma': 1.0,},
).fit(X[:, np.newaxis], Y[:, np.newaxis])

hsic_clf.score(X), hsic_clf.score(X, normalize=True)
```
